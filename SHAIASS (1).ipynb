#TASK 1: 
Basic Data Exploration: Identify the number of rows and columns in the dataset, determine the data types of each column, and check for missing values in each column.

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Here i used read_csv function from pandas library to make a data frame:
salaries_df = pd.read_csv("Salaries.csv")
```

```python
#To Identify the number of rows and columns shape function will give us the number of (rows,columns)
df_shape=salaries_df.shape
print('(rows , columns):',df_shape)
```

```python
#Applying the dtype attribute to determine the data types of each column
dt_types = salaries_df.dtypes
print(dt_types)
```

```python
#To check if there are missing values in each column, I can use two ways:
#First one: BasePay, OvertimePay, OtherPay,Benefits 
print('Dataset Info:',salaries_df.info())
```

```python
#Second way:
missing_values = salaries_df.isnull()
print('Missing values in each column:', missing_values)
```

TASK 2: Descriptive Statistics: Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation

```python
#For calculating some statistical values of the DataFrame i use describe function.
salaries_df.describe()
```

```python
#If I'm going to calculate these values for specific column like (salary):
print("Max:", salaries_df['TotalPay'].max())
print("Min:", salaries_df['TotalPay'].min())
print("Central tendency:")
print("Median:", salaries_df['TotalPay'].median())
print("Mean:", salaries_df['TotalPay'].mean())
print("Mode:", salaries_df['TotalPay'].mode())
print("Variance:", salaries_df['TotalPay'].var())
print("std:",math.sqrt( salaries_df['TotalPay'].var()))
percentiles = np.percentile(salaries_df['TotalPay'], [25, 50, 75])
print("Percentiles [25th, 50th, 75th]:", percentiles)
```

```python
#To determine the range of salaries:
salaries_range= salaries_df['TotalPay'].max() - salaries_df['TotalPay'].min()
print(salaries_range)
```

TASK 3: Data Cleaning: Handle missing data by suitable method with explain why you use it.

from task 1 I noticed that these columns have a null or missing values: 
BasePay, OvertimePay, OtherPay,Benefits.
the commen thing between these columns is that they all have the same numerical data type (float64), and that is mean i can solve this problem in different ways:
1. the attribute mode.
2. the attribute mean.

but because we have numerical values the second choice will be better 

```python
missingValuesColumns = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']

#Here I make a for loop to fill missing values with mean method
for i in missingValuesColumns:
    salaries_df[i].fillna(salaries_df[i].mean(), inplace=True)
    
#To make sure that I fill missing values:
salaries_df.info()
```

TASK 4: Basic Data Visualization: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
There are a lot of ways to visualize a salary column, such as:

1.Box plot

2.Distribution plot

3.histograms
```

```python
#Box plot:it shows how the data is distributed and it also shows any outliers.
sns.boxplot(x=salaries_df['TotalPay'], color='purple')
plt.title('Boxplot of TotalSalary')
plt.show()
print("\n")
```

```python
#Distribution plot:
sns.histplot(salaries_df['TotalPay'], kde=True, color='purple', linewidth=2)
plt.title('Distribution of Salary')
plt.show()
print("\n")
```

```python
#histograms: represents the frequency of Numerical Data using bars:
salaries_df['TotalPay'].plot(kind='hist', bins=20, color='black', title=f'Histogram of Salary column')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group by job titles and count the number of employees in each job title
job_title_counts = salaries_df['JobTitle'].value_counts()

# Choose the top N job titles to include in the pie chart
top_job_titles = job_title_counts.head(10)

# Plot a pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_job_titles, labels=top_job_titles.index, autopct='%1.2f%%', startangle=140)
plt.title('Proportion of Employees in Different Job Titles')
plt.show()
```

TASK 5: Grouped Analysis: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

```python
#Group by jobtitle column:
jobtitle_group = salaries_df.groupby(['JobTitle'])

#To calculate the mean 'TotalPay' for each job title group
group_mean = jobtitle_group['TotalPay'].mean()
group_mean
```

```python
#To calculate summary statistics for each group, I take the JobTitle as a sample:
salaries_df.groupby('JobTitle').describe()
```

TASK 6: Simple Correlation Analysis: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

```python
#I choose 'TotalPay' and 'TotalPayBenefits' columns to calculate the correlation between them
correlation_value = salaries_df['TotalPay'].corr(salaries_df['TotalPayBenefits'])
correlation_value
```

```python
#Scatter plot to visualize the relatrion between 'TotalPay' and 'TotalPayBenefits':
plt.scatter(salaries_df['TotalPay'], salaries_df['TotalPayBenefits'])
plt.xlabel('TotalPay')
plt.ylabel('TotalPayBenefits')
plt.title('TotalPay vs TotalPayBenefits')
plt.show()
```
